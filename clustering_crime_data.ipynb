```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('crime_data.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data
```

```python
array=data.values
array
```

```python
# we need stardize the values
crime=StandardScaler().fit_transform(array)
crime
```

```python
# we need find epsilon value for db scan
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(crime)
distances, indices = nbrs.kneighbors(crime)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
# as you can see the curve is around 1 so eps =1
# dbscan clustering
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
data2=data.copy()
data2['db_cluster']=dbscan.labels_
data2
```

```python
data2.groupby('db_cluster').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(data2['db_cluster'],data2['UrbanPop'], c=dbscan.labels_)
```

```python
# as you can see there are three clusters
```

```python
# hierarchical clustering
```

```python
data3=data.copy()
data3
```

```python
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
trans_data = pd.DataFrame(trans.fit_transform(data3))
trans_data
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram
#p = np.array(df_norm) # converting into numpy array format
z = linkage(trans_data,  method="complete",metric="euclidean")
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram( z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(trans_data)
```

```python
cluster_labels=pd.Series(h_complete.labels_)
data2['h_clust']=cluster_labels # creating a  new column and assigning it to new column
data2
```

```python
# K MEANS CLUSTERING
from sklearn.cluster import KMeans
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(trans_data)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
# as you can see the curve so, we can say k=5
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(trans_data)
```

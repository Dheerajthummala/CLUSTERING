```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
data
```

```python
data.info()
```

```python
data2=data.drop(['ID#'],axis=1)
data2
```

```python
data2.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
trans_data = pd.DataFrame(trans.fit_transform(data2))
trans_data
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram
#p = np.array(df_norm) # converting into numpy array format
z = linkage(trans_data,  method="complete",metric="euclidean")
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram( z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(trans_data)

cluster_labels=pd.Series(h_complete.labels_)
data['h_clust']=cluster_labels # creating a  new column and assigning it to new column
data
```

```python
data.groupby('h_clust').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(data['h_clust'],data['Qual_miles'], c=cluster_labels)
```

```python
# DB SCAN CLUSTERING
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data2
```

```python
array=data2.values
array
```

```python
# we need stardize the values
stan_data=StandardScaler().fit_transform(array)
stan_data
```

```python
# we need find epsilon value for db scan
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=6)
nbrs = neigh.fit(stan_data)
distances, indices = nbrs.kneighbors(stan_data)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
# as you can see the curve is around 1 so eps =1
# dbscan clustering
```

```python
dbscan=DBSCAN(eps=1,min_samples=6)
dbscan.fit(stan_data)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
data['db_cluster']=dbscan.labels_
data
```

```python
data.groupby('db_cluster').agg(['mean']).reset_index()
```

```python
clust=pd.DataFrame()
clust['award']=data['Award?']
clust['h_pred']=data['h_clust']
clust['db_pred']=data['db_cluster']
clust
```

```python
# K MEANS CLUSTERING
from sklearn.cluster import KMeans
```

```python
trans_data
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(trans_data)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
# as you can see the curve so, we can say k=5
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(trans_data)
```

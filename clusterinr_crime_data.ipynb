```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('crime_data.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data
```

```python
array=data.values
array
```

```python
# we need stardize the values
crime=StandardScaler().fit_transform(array)
crime
```

```python
# we need find epsilon value for db scan
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(crime)
distances, indices = nbrs.kneighbors(crime)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
# as you can see the curve is around 1 so eps =1
# dbscan clustering
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
data['cluster']=dbscan.labels_
data
```

```python
data.groupby('cluster').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(data['cluster'],data['UrbanPop'], c=dbscan.labels_)
```
